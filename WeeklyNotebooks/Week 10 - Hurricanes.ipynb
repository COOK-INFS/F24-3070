```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://www.nhc.noaa.gov/data/hurdat/hurdat2-1851-2023-051124.txt')

df.head()
```

```python
df.info()
```

```python
# Define some column names while importing the data file.

column_names = [
    'ID-Date', 'Time', 'RecordID', 'Status', 'Latitude', 'Longitude',
    'WindSpeed', 'Pressure','ExtraField1', 'ExtraField2', 'ExtraField3',
    'ExtraField4', 'ExtraField5', 'ExtraField6', 'ExtraField7', 'ExtraField8',
    'ExtraField9', 'ExtraField10', 'ExtraField11', 'ExtraField12'
]

# Read the file, restrict it to the number of columns and column names we want to use.
df = pd.read_csv('dataFiles\\hurdat2-1851-2023-051124.txt', names=column_names, usecols=range(len(column_names)), on_bad_lines='skip')

df.head()
```

## Extract the ID, Year, and Month from our ID-Date column

```python
# Create new columns in our dataFrame with empty values for Year and Month
df['YEAR'] = ''
df['MONTH'] = ''

# Only extract the Year and Month where 'ID-Date' does not contain 'AL'
# Extract the Year
df.loc[~df['ID-Date'].str.contains('AL'), 'YEAR'] = df['ID-Date'].str[:4]

# Extract Month
df.loc[~df['ID-Date'].str.contains('AL'), 'MONTH'] = df['ID-Date'].str[4:6]

# Create a Unique ID column that copies the AL identifier
df['UNIQUE_ID'] = df['ID-Date'].apply(lambda x: x if 'AL' in x else None)

# To copy this to the entire dataFrame we're going to use 'Forward-Fill'
df['UNIQUE_ID'] = df['UNIQUE_ID'].ffill()

df.head()
```

# Number of Hurricanes by Year

```python
hurricanes_by_year = df.groupby('YEAR')['UNIQUE_ID'].count().reset_index(name='COUNT')

hurricanes_by_year
```

```python
hurricanes_by_year.info()
```

```python
# Visualize the number of Storms

# Convert the Year column to numeric, coerce errors to NaN
df['YEAR'] = pd.to_numeric(df['YEAR'], errors='coerce')

# Drop rows where YEAR is NaN (This is the unique ID rows)
df_clean = df.dropna(subset=['YEAR'])

# Convert Year to an Integer using .loc
df_clean.loc[:, 'YEAR'] = df_clean['YEAR'].astype(int) # The [:, 'YEAR] instructs Pandas to select all rows in the 'YEAR' column.

# Now group our cleaned data
hurricanes_by_year = df_clean.groupby('YEAR')['UNIQUE_ID'].nunique()

# Visualize
plt.figure(figsize=(10,5))
plt.plot(hurricanes_by_year.index, hurricanes_by_year.values, marker='o')

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Hurricanes by Year')

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(25)) # Controls the interval of the X axis values.
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
# Visualize the number of Storms

# Convert the Year column to numeric, coerce errors to NaN
df['YEAR'] = pd.to_numeric(df['YEAR'], errors='coerce')

# Drop rows where YEAR is NaN (This is the unique ID rows)
df_clean = df.dropna(subset=['YEAR'])

# Convert Year to an Integer using .loc
df_clean.loc[:, 'YEAR'] = df_clean['YEAR'].astype(int) # The [:, 'YEAR] instructs Pandas to select all rows in the 'YEAR' column.

# Now group our cleaned data
hurricanes_by_year = df_clean.groupby('YEAR')['UNIQUE_ID'].nunique()

# Calculate the cumulative average of Hurricanes
cummulative_avg_hurricanes = hurricanes_by_year.expanding().mean()

# Visualize
plt.figure(figsize=(10,5))
plt.plot(hurricanes_by_year.index, hurricanes_by_year.values, marker='o')

# Plot an average line
plt.plot(cummulative_avg_hurricanes.index, cummulative_avg_hurricanes.values, color='red', linestyle='--', label='Cumulative Average')

plt.xlabel('Year')
plt.ylabel('Number of Hurricanes')
plt.title('Hurricanes by Year')

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(25)) # Controls the interval of the X axis values.
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

# Hurricanes by Category

Here we introduce the 'Functions' within Python

```python
# Create a function to defince category based on wind speed
def categorize_wind_speed(speed):
    if 74 <= speed <= 95:
        return "Category 1"
    elif 96 <= speed <=110:
        return "Category 2"
    elif 111 <= speed <=129:
        return "Category 3"
    elif 130 <= speed <156:
        return "Category 4"
    elif speed >= 157:
        return "Cateogry 5"
    
# Create a new column in our dataframe called 'CATEGORY' and populate it with the Category based on the Windspeed.
df['CATEGORY'] = df['WindSpeed'].apply(categorize_wind_speed)


# Group our data
hurricanes_by_year_category = df.groupby(['YEAR', 'CATEGORY'])['UNIQUE_ID'].count().unstack(fill_value=0)


# Plot the graph
plt.figure(figsize=(12,6))

# Loop through the data to read every row and show a line on the plot for each category of hurricane.
for category in hurricanes_by_year_category.columns:
    plt.plot(hurricanes_by_year_category.index, hurricanes_by_year_category[category], label='category')


plt.xlabel("Year")
plt.ylabel("Number of Hurricanes")
plt.title("Number of Hurricanes per year and category")
plt.grid()
plt.legend()
plt.show()
```

```python
hurricanes_by_year_category.plot(kind="barh", stacked=True, figsize=(10,40))

# Plot the graph
plt.ylabel('Year')
plt.xlabel('Number of Hurricanes')
plt.title('Number of Hurricanes by Year and Category')
plt.grid(axis='x')
plt.legend()

plt.show()
```
